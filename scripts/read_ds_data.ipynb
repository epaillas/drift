```python
from lsstypes import read

data = read('/Users/epaillas/code/drift/outputs/dsg_measured.hdf5')

rebin = 5
data = data.select(k=slice(0, None, rebin))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(4, 3))

for qid in range(4):
    print(f"Quantile {qid}:")

    quantile = data.get(quantiles=qid)

    k = quantile.get(ells=0).coords("k")
    p0 = quantile.get(ells=0).value()

    ax.plot(k, k ** 2 * p0)
ax.set_xlabel('k [h/Mpc]')
ax.set_ylabel('k^2 P0 [Mpc/h]^3')
ax.legend()
ax.set_xscale('log')
plt.show()

```
